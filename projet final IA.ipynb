# changing scheduled day and Appointment Day

```python
df1 = data[['ScheduledDay','AppointmentDay']]
```

```python
df1
```

```python
def delete_letters (a):
    y=(a.replace('T', ' '))
    z=(y.replace('Z', ''))
    return (z)    
newdf = df1.applymap(delete_letters)
```

```python
newdf
```

```python
newdf2=newdf.copy()
```

```python


newdf['ScheduledDay1'] = pd.to_datetime(newdf['ScheduledDay'])
newdf['AppointmentDay1'] = pd.to_datetime(newdf['AppointmentDay'])

```

```python
newdf
```

```python
newdf['Scheduled_date'] = newdf['ScheduledDay1'].dt.date
newdf['Appointment_date'] = newdf['AppointmentDay1'].dt.date
newdf['Wait_Days'] = (newdf['Appointment_date']-newdf['Scheduled_date']).dt.days
newdf['Wait_Days']=newdf['Wait_Days'].astype(int)
```

```python
newdf
```

```python
import datetime
import calendar
```

```python
def date2weekday(date_str):
    my_date= datetime.datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S').date()
    day = calendar.day_name[my_date.weekday()]
    return day
```

```python
newdf1=newdf2.applymap(date2weekday)
```
